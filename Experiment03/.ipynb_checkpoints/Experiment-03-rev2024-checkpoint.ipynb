### Model Fitting

For this round, I'll try using initial parameters based on

$$V(t)=RCV_0 \omega_0 \exp(-\gamma t/2)\cos(\omega_0 t + \phi_0),$$

so the guess for the amplitude would be

$$A = RCV_0 \omega_0 = \frac{RCV_0}{\sqrt{LC}},$$

the time constant would be

$$\tau = 2/\gamma = L/R,$$

and the resonant frequency would be

$$f = \omega/2\pi = 1/(2\pi \sqrt{LC}).$$

Lastly, I'll just assume a standard phase shift of $\phi_0 = \pi/2$

```python
def amplitude_guess(capacitance, volt_amplitude):
    return resistance * capacitance * volt_amplitude / \
        np.sqrt(inductance * capacitance)

TIME_CONSTANT_GUESS = inductance / resistance

def frequency_guess(capacitance):
    return 1 / (2 * np.pi * np.sqrt(inductance * capacitance))

PHASE_GUESS = np.pi/2
```

```python
def standard_guess_model_1(capacitance, volt_amplitude=8.5):
    amp=amplitude_guess(capacitance_2, volt_amplitude)
    freq=frequency_guess(capacitance_2)
    
    print('Guessing amplitude: %.3e' % amp)
    print('Guessing time constant: %.3e' % TIME_CONSTANT_GUESS)
    print('Guessing resonant frequency: %.3e' % freq)
    print('Guessing phase shift: %.3e' % (np.pi/2))
    
    return RingdownFitParameters(
        amplitude=amp,
        time_constant=TIME_CONSTANT_GUESS,
        resonant_frequency=freq,
        phase=np.pi/2
    )
```

```python
model2_1 = standard_guess_model_1(capacitance_2)
```

```python
results2_1 = autofit(packed_data2, gopts, model2_1)
results2_1.tabulate()
```

I'll switch to using a standard guess model based on the period, $2$ $\mu s$, and I'll just manually guess the time constant.

```python
def standard_guess_model_2(period, tau, volt_amplitude=8.5):
    freq=1/period
    
    print('Guessing amplitude: %.3e' % volt_amplitude)
    print('Guessing time constant: %.3e' % tau)
    print('Guessing resonant frequency: %.3e' % freq)
    print('Guessing phase shift: %.3e' % (np.pi/2))
    
    return RingdownFitParameters(
        amplitude=volt_amplitude,
        time_constant=tau,
        resonant_frequency=freq,
        phase=np.pi/2
    )
```

```python
def plot_capacitance_versus(values, label, units=None):
    plt.figure()
    plt.title(f'{label} vs. Capacitance')
    plt.plot(capacitance_values,values)
    plt.xlabel('Capacitance (F)')
    plt.ylabel(f'{label} %s' % ('(%s)' % units if units else ''))
    plt.show()
```

**Observation:** The best plots I got, by far, were at lower capacitances. My guess is that this could be due to the charge buildup in the capacitor causing a lower voltage drop across the resistor, making the oscilloscope have to make more precise measurements, perhaps? I'm not certain.
